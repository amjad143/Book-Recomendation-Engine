```python
# Merge this books data set to the ratings columns
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
# Checking the reviews of the each book
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
# Mearging this dataset on the ratings with books data set
final_rating = ratings_with_books.merge(number_rating, on='title')
final_rating
```

### Accepting more than 50 datasets we selected

# Creating a Pivot Table

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

```python
# Filling missing values 
book_pivot.fillna(0,inplace=True)
book_pivot
```

# Converting into Sparesmmatrix
* This means it wont considerd zero values

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

## Model deployment
* Clustring Model

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

* Slecting single book

```python
# Example of selecting a book number
book_pivot.index[237]
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

* 0. means independent of that book and remaining are the id

```python
suggestions
```

* These are the books that actuaaly recomending

```python
# To display their names
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
# To Create Recomend Function
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print("The suggestion for",book_name,'are :')
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
